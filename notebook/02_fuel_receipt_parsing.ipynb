```python
import re
import pytesseract
from PIL import Image
```

```python
# OCR the image
text = pytesseract.image_to_string(
    Image.open("../xraw-data/fuel-receipts/01f.png"),
    config="--psm 6" #single uniform block of text
)
print(text)
```

```python
# Find club manager
# Handles no club manager listed
club_manager = re.search(r'CLUB\s+MANAGER(?:[ ]([A-Z]+))?', text)

if club_manager:
    print(club_manager)
    print(club_manager.group(1))
```

```python
# Find date in format mm/dd/yy hh:mm
date = re.search(r'\d{2}/\d{2}/\d{2}\s+\d{2}:\d{2}', text)

if date:
    print(date)
    print("date group:", date.group(0))
```

```python
# Find location in format CITY, ST
# Handles multi-word city names 
location = re.search(r'[A-Z]+(?:[][A-Z]+)*,\s+[A-Z]{2}', text)

if location:
    print(location)
    print("location group:", location.group(0))
```

```python
# Find pump number and fuel type

pump_fuel = re.search(r'Pump#\s+(\d+)\s+([A-Z]+)', text)

if pump_fuel:
    print(pump_fuel)
    print("pump number:", pump_fuel.group(1))
    print("fuel type:", pump_fuel.group(2))
```

```python
# Find gallons in format Gallons 12.345
gallons = re.search(r'Gallons\s+(\d+\.\d{3})', text)

if gallons:
    print(gallons)
    print("gallons:", gallons.group(1))
```

```python
# Find price per gallon in format Price/Gal $1.234
price_per_gallon = re.search(r'Price/Gal\s+\$(\d+\.\d{3})',text)

if price_per_gallon:
    print(price_per_gallon)
    print("price per gallon:", price_per_gallon.group(1))
```

```python
# Find total cost in format Fuel Sale $12.34

total_cost = re.search(r'Fuel Sale\s+\$(\d+\.\d+)', text)

if total_cost:
    print(total_cost)
    print("total:", total_cost.group(1))
```

```python
# Find payment method (VISA CREDIT)
paymentMethod = re.search(r'(VISA|MASTERCARD)\s+(CREDIT|DEBIT)',text)

if paymentMethod:
    print(paymentMethod)
    print("card:", paymentMethod.group(1))
    print("type:", paymentMethod.group(2))
```
